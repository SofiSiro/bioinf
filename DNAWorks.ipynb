# DNAWorks input
Here you can see the widgets to fill the file easily. Just run the cell and put params what you need. In most cases you need only lenght, temperature and nucleotide seq.

Output of program is file with input data for DNAWorks. Use the file for program running.

```python
import ipywidgets as widgets
from IPython.display import display

# Функция для форматирования нуклеотидной последовательности
def format_sequence(sequence):
    sequence = sequence.replace(" ", "").replace("\n", "")
    lines = []
    for i in range(0, len(sequence), 60):
        line = sequence[i:i+60]
        formatted_line = ' '.join(line[j:j+10] for j in range(0, len(line), 10))
        lines.append(formatted_line)
    return '   ' + '\n   '.join(lines)

# Функция для создания файла .inp
def create_inp_file(params):
    filename = f"{params['title']}.inp"
    with open(filename, 'w') as file:
        if params['title']:
            file.write(f"title {params['title']}\n")
        if params['melting_low'] or params['melting_high'] or params['melting_tolerance']:
            file.write(f"melting low {params['melting_low']} high {params['melting_high']} tolerance {params['melting_tolerance']}\n")
        if params['length_low'] or params['length_high']:
            file.write(f"length low {params['length_low']} high {params['length_high']}\n")
        if params['previous_file']:
            file.write(f"previous {params['previous_file']}\n")
        if params['pattern']:
            file.write("pattern\n")
            file.write(params['pattern'] + "\n")
        if params['codon']:
            file.write(f"codon {params['codon']}\n")
        if params['protein']:
            file.write("protein\n" + params['protein'] + "\n")
        if params['nucleotide']:
            file.write("nucleotide\n" + format_sequence(params['nucleotide']) + "\n")
        # Добавляем \\ автоматически
        file.write("//\n")
    print(f"Файл '{filename}' создан.")
    return filename

# Поля формы
title_input = widgets.Text(description="Title:", layout=widgets.Layout(width='300px'))
melting_low_input = widgets.Text(description="Melting Low:", layout=widgets.Layout(width='300px'))
melting_high_input = widgets.Text(description="Melting High:", layout=widgets.Layout(width='300px'))
melting_tolerance_input = widgets.Text(description="Melting Tol.:", layout=widgets.Layout(width='300px'))
length_low_input = widgets.Text(description="Length Low:", layout=widgets.Layout(width='300px'))
length_high_input = widgets.Text(description="Length High:", layout=widgets.Layout(width='300px'))
previous_file_input = widgets.Text(description="Previous File:", layout=widgets.Layout(width='300px'))
pattern_input = widgets.Textarea(description="Pattern:", layout=widgets.Layout(width="300px", height="100px"))
codon_input = widgets.Text(description="Codon:", layout=widgets.Layout(width='300px'))
protein_input = widgets.Textarea(description="Protein:", layout=widgets.Layout(width="300px", height="100px"))
nucleotide_input = widgets.Textarea(description="Nucleotide:", layout=widgets.Layout(width="300px", height="100px"))

# Кнопка
button = widgets.Button(description="Создать файл .inp")

# Обработчик кнопки
def on_button_click(b):
    params = {
        "title": title_input.value,
        "melting_low": melting_low_input.value,
        "melting_high": melting_high_input.value,
        "melting_tolerance": melting_tolerance_input.value,
        "length_low": length_low_input.value,
        "length_high": length_high_input.value,
        "previous_file": previous_file_input.value,
        "pattern": pattern_input.value,
        "codon": codon_input.value,
        "protein": protein_input.value,
        "nucleotide": nucleotide_input.value
    }
    create_inp_file(params)

button.on_click(on_button_click)

# Отображение формы
display(title_input, melting_low_input, melting_high_input, melting_tolerance_input,
        length_low_input, length_high_input,
        previous_file_input, pattern_input, codon_input, protein_input, nucleotide_input, button)
```
